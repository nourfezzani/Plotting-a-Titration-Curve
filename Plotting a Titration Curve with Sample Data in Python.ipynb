```python
#Importing the array processing package
import numpy 
import os
```

```python
#This is the format we want to use for processing our data
help(numpy.genfromtxt)
```

```python
#Loading our data into the notebook. We use 'data' to indicate the folder that the .CSV file is located in. We write 'Sample_Titration_data1.csv' to indicate the file that our data is located in.
titration_data=os.path.join('data','Sample_Titration_data1.csv')

#This line tells Python to import our data using the format noted previously. Print data tells Python to show us what we have coded so far.
data=numpy.genfromtxt(fname=titration_data, delimiter=',', dtype='unicode', skip_header=1)
print(data)
```

```python
#Separating the headers (title of each column) from the values.
headers = data[0]
print(headers)
```

```python
#Slicing the data to get the numerical values only (without the header). 
data = data[1:]
print(data)
```

```python
#Float changes all the entries in the array to numbers. We are giving our data a new name - alldata - so we do not get confused.
alldata = data.astype(numpy.float)
print(alldata)
```

```python
#We are importing the plotting library for Python and renaming it plt to make it easier to type its commands.
import matplotlib.pyplot as plt
```

```python
#Now we are telling Python to plot the data we imported above.
plt.figure()

#This tells Python the x values are the numbers in alldata. The : tells python to use all the values and the 0 specifies it to use the ones in the first column.
x=alldata[:,0]
#This tells Python the y values are the numbers in alldata. The : tells python to use all the values and the 1 specifies it to use the ones in the second column.
y=alldata[:,1]

#Finally, this command tells Python to label x as Volume and y as pH, and tells Python to plot the x and y values noted above.
plt.xlabel('Volume Added (mL)')
plt.ylabel('pH')
plt.plot(x,y)
```

The data given was the titration of a weak base with a strong acid. The acid and base reacted in this titration to form an acidic solution (which has a pH lower than 7), as seen in the plot of the curve.

The equivalence point, which is the point at which there is enough acid and base to neutralize the solution, for this example occurred at a pH of ~4.4. After adding ~4.75 mL titrant acid, the solution reached complete neutralization. 

For this example, the pH at the equivalence point was lower than 7 (neutral) since we used a strong acid and weak base in the titration. The strong acid used resulted in the formation of the conjugate acid of the weak base, which reacted with water to make the solution acidic. If we had used a strong base and a strong acid, the pH at the equivalence point would be equal to 7.
